# OSM with Overpass API

## Basic 

```python
import requests
```

```python
%%time
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = """
[out:json];
area["ISO3166-1"="PT"][admin_level=2];
node["place"="city"](area);
out;
out count;
"""
data = requests.get(overpass_url, params={'data': overpass_query}).json()
```
