```python
import pandas as pd
```

```python
imputed_data = pd.read_csv('https://raw.githubusercontent.com/McNickSisto/world_data_league/main/stage_final/data/Imputed_Data_Final.csv')
imputed_data = imputed_data.drop(columns = ['Unnamed: 0'])
complaints = pd.read_csv('https://raw.githubusercontent.com/McNickSisto/world_data_league/main/stage_final/data/complaints_2018.csv')
```

```python
imputed_data.head(2)
```

```python
complaints.head(2)
```

```python
df = complaints.groupby(['Data'])['Unnamed: 0'].count().reset_index(
  name='Count').sort_values(['Data'], ascending=True)
```

```python
df
```

```python
df_final_comp = imputed_data.merge(df, left_on='date', right_on='Data', how ='left')
df_final_comp = df_final_comp.drop(columns = ['Data'])
df_final_comp = df_final_comp.fillna(value={'Count' : 0})
df_final_comp = df_final_comp.rename(columns = {'Count': 'Complaints_no'})
df_final_comp.head(30)
```

```python
df_final_comp['date_hour'] = pd.to_datetime(df_final_comp['date_hour'])
df_final_comp['date_hour']= df_final_comp['date_hour'].dt.strftime("%d-%m-%y %H:%M")
```

```python
df_open_count = pd.read_csv('https://raw.githubusercontent.com/McNickSisto/world_data_league/main/stage_final/data/opening_count_2018.csv')
df_open_count = df_open_count.drop(columns= ['Unnamed: 0'])
df_open_count['hour'] = pd.to_datetime(df_open_count['hour'] )
df_open_count['hour']  = df_open_count['hour'] .dt.strftime("%d-%m-%y %H:%M")
df_open_count = df_open_count.rename(columns = {'hour': 'time_hour'})
df_open_count.head(2)
```

```python
df_final = df_final_comp.merge(df_open_count, left_on='date_hour', right_on ='time_hour', how = 'left')
df_final
```

```python
df_final = df_final.drop(columns = ['time_hour'])
df_final
```
