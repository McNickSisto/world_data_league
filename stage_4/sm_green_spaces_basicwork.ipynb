# Research
Researchers are still working out exactly why green space is so beneficial, but it clearly provides health benefits across the population. It can encourage exercise, provide spaces for socializing, decrease noise and air pollution, and improve immune function by providing exposure to beneficial microbiota.

**Benifits of spending time at green spaces:**

It also can help with psychological restoration; that is, green space provides a respite for over-stimulated minds.Exposure to green space is comparable to family history and parental age when predicting mental health outcomes. Only socioeconomic status was a slightly stronger indicator.Green space most strongly protects against mood disorders, depression, neurotic behavior, and stress-related issues, the study found, signaling that psychological restoration may be the strongest protective mechanism that green space offers.The effect of green space is also dose-dependent, meaning those who have longer exposures to green space have greater mental health benefits.

Green spaces and other nature-based solutions offer innovative approaches to increase the quality of urban settings, enhance local resilience and promote sustainable lifestyles, improving both the health and the well-being of urban residents.Urban green space is an important investment that local authorities can make on behalf of citizens and their well-being.

Green space interventions are defined as actions that significantly modify the quality, quantity and accessibility of green space. This can be done by establishing new green spaces or by changing the characteristics and functions of existing ones.

**A broad spectrum of intervention types can be implemented at different scales in private or public spaces.
These include:**

*   roadside greenery and vegetation barriers along streets or rail tracks parks and urban meadows

*   facilitated access to urban woodlands,forests and natural wildlife areas
*   small urban green spaces (such as gardens or pocket parks) and playgrounds
*   green roofs and facades
*   greenways and corridors (such as green trails for walking/cycling)
*   coastal, riverside or lakeside trails, linking green with blue spaces
*   recreational and urban gardening facilities(such as community gardens, sport and play areas and school grounds)

**Opportunities to involve urban green space interventions in urban planning include:**

*   development of new residential neighbourhoods,community facilities,business parks or transport infrastructure projects
*   regeneration projects and urban renewal initiatives
*   brownfield development and rehabilitation of industrial areas
*   urban gardening/agriculture projects

**Factors effecting the use of green spaces**

Demographic characters (e.g., age, occupation, culture, lifestyle, household size, experience, and leisure time), interfering variables (e.g., poor behavior, leisure capacity, and resource impact), behavioral preferences (e.g., recreation motivation, expectation value, costs, and accessibility), and recreation needs (e.g., landscape feature, facility, and management).

Moreover, other impact factors in urban green space might affect perception of recreational values and satisfaction levels, such as biodiversity, littering, accessibility, usability, living context, quality of vegetation, and infrastructure, as well as gender, education, monthly income, and dwelling location of residents.

Mean distance to green space increased with neighbourhood deprivation.Additionally, green spaces in the more deprived neighbourhoods presented significantly more safety concerns, signs of damage, lack of equipment to engage in active leisure activities, and had significantly less amenities such as seating, toilets, caf√©s, etc.

The  attractiveness  concept  is  associated  with  specific characteristics  of several  urban green  space categories,  which differ from the context of the surrounding area, defining its use.

It is considered by researches that there are "a set of factors that influence the use of public green spaces,  including:  the  quality  and  quantity  of  spaces,  the sociodemographic  characteristics of  potential users,  access to facilities, ability to  meet the  needs of users,  maintenance and the perception of safety". In this sense, the choices of the space do not fall only on associated physical factors, but may also be influenced  by  individual  preferences, even though it is expected to visit the spaces that are nearby the residence or work places, or those which have higher levels of attractiveness associated with quality.
```python
#Let's check the data dictionary to have a better understanding on the dataset 
green_spaces=pd.read_csv('/content/WDL-Stage4/Data/Attracting Population/WDL_Green_Areas_Dataset_V1.txt',sep="|",encoding='latin-1')
green_spaces.head(4)
```

```python
#Let's check if there are any null values
green_spaces.info()
#There are no null values in the dataset
```

```python
#Let's check if all the Green_Zone_names are unique 
print("No.of.green_zones : {0}".format(len(set(green_spaces['Green_Zone_name']))))
#It seems all the green zone names are not unique and the unique values are half the total no.of values present
#There might be chance for each green_zone information repeated twice\
#Let's check if all the green spaces are repeated only twice
print("count of green spaces that are repeated twice : {0}".format(sum(green_spaces['Green_Zone_name'].value_counts()==2)))
```

```python
green_spaces.head(2)
```

It is observed that each green_zone is repeated twice, one with the resident information and the other with non-resident information.

In total we have information about 64 green zones in the given dataset.

All the columns with respect to resident and non-resident of a partiuclar green_zone are having the same values except Daily average of people in influence area.





```python
#Let's check if the average people visting green spaces are mostly residents or the non-residents
green_spaces.groupby('Resident')['Daily average of people in influence area'].describe()
```

If taken as whole, the average of people in the influnece area are mostly non-residents. 

```python
#Let's try working on resident data
res_green_spaces=green_spaces[green_spaces['Resident']==1]
res_green_spaces.head(2)
```

```python
#Graph to analyse the numerical distribution of data values
res_green_spaces.plot(figsize=(50,20),grid=True,fontsize=25)
plt.legend(fontsize=20)
```

```python
#Let's check which factors effect the most on people visting the green zones
corr_resident = res_green_spaces.corr()
f, ax = plt.subplots(figsize=(14, 12))
sn.heatmap(corr_resident, annot = True )
plt.show()
```

```python
corr_resident['Daily average of people in influence area'].sort_values(ascending=False)
#This shows how the attributes are related to people in the influence area
```

The Daily average of people in influence area is most influenced by Number of people resident in influence area and Number of families in influence area and is least influenced by percentage of senior people, percentage of population until 19 years old, and number of museums in influence area.

```python
sn.jointplot(x = 'Number of people resident in influence area', y = 'Daily average of people in influence area', data = res_green_spaces, kind = 'reg');
```
