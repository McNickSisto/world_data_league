```python
import pandas as pd
import geopandas as gpd
import folium
from ipywidgets import interact, interactive, fixed, interact_manual, IntSlider
import ast
from shapely import wkt
from shapely.geometry import Point
```

```python
df = pd.read_csv('raw_data/routes_usage1.csv', index_col=0, converters={'geometry':Point})
# gdf = gpd.GeoDataFrame(df).set_geometry('geometry')
# df['geometry'] = df['geometry'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(df, crs='epsg:4326')
```

```python
gdf.geometry
```

```python
df = df[df.geometry == df.geometry]
```

```python
df
```

```python
def mapa(hour):
    m = folium.Map([6.25, -75.58], zoom_start=13)
    hour_df = df[df[str(hour)] == df[str(hour)]][[str(hour), 'geometry']]
    hour_df = hour_df.head(50)

# add tooltip
    for c, row in hour_df.iterrows():
        try:
            folium.Choropleth(row.geometry, data=row, columns=[str(hour), 'geometry'], line_weight=5).add_to(m)
        except Exception as e:
            print(f'couldnt: {c}, erroe: {e}')

    return m

interact(mapa, hour=IntSlider(min=0, max=23, step=1, value=8))
```
